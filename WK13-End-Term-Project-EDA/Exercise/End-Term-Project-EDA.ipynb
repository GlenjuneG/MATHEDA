# Data Cleaning - Droping NaN Values

```python
df_w.isna().sum()
```

```python
# Culmen Depth (mm)
df_w = df_w.dropna(subset=['Culmen Depth (mm)'])
df_w.isna().sum().reset_index()
```

```python
# Flipper Length (mm)
df_w = df_w.dropna(subset=['Flipper Length (mm)'])
df_w.isna().sum().reset_index()
```

```python
# Body Mass (g)
df_w = df_w.dropna(subset=['Body Mass (g)'])
df_w.isna().sum().reset_index()
```

```python
# Sex (Male/Female)
df_w = df_w.dropna(subset=['Sex (Male/Female)'])
df_w.isna().sum().reset_index()
```
